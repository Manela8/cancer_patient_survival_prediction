# Data_Cleaning

### Dataset : Breast_Cancer_METABRIC

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Load dataset
PROJECT_ROOT = Path("..").resolve()
DATA_DIR = PROJECT_ROOT / "data"
RAW_DATA_PATH = DATA_DIR / "Breast_Cancer_METABRIC.csv"
CLEAN_DATA_PATH = DATA_DIR / "cleaned_data.csv"
print("Root Directory: ", PROJECT_ROOT)
print("Data Directory: ", DATA_DIR)
print("Raw Data Path: ", RAW_DATA_PATH)
print("Cleaned Data Path: ", CLEAN_DATA_PATH)
```

```python
# Load data
data = pd.read_csv(RAW_DATA_PATH)
data.head()
```

```python
# Data back-up
df = data.copy()
```

```python
# Meta data
print("No. of rows: ", df.shape[0])
print("\nNo. of columns: ", df.shape[1])
print("\nAll columns: \n", df.columns.to_list())
print("\nDuplicate Count: ", df.duplicated().sum())
print("\nNull Values: \n", df.isnull().sum())
print("\nInfo:\n")
df.info()
```

```python
# Drop irrelevant columns
df.drop(columns = ["Patient ID", "Cancer Type", "ER status measured by IHC", "HER2 status measured by SNP6"], axis = 1, inplace = True)
print("Final list of columns: \n", df.columns.to_list())
```

```python
# Save c;lean data
df.to_csv(CLEAN_DATA_PATH, index = False)
```
